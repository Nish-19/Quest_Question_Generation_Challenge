```python
import pandas as pd
import spacy
```

```python
nlp = spacy.load('en_core_web_sm')
```

```python
train_df = pd.read_csv('original/train.csv')
```

```python
# local_or_sum and attribute 
co_mat_ls_a = pd.crosstab(train_df.local_or_sum, train_df.attribute1)
co_mat_ls_a
```

```python
# local_or_sum and ex_or_im
co_mat_ls_ei = pd.crosstab(train_df.local_or_sum, train_df.ex_or_im)
co_mat_ls_ei
```

```python
# attribute1 and ex_or_im
co_mat_attr_ex = pd.crosstab(train_df.attribute1, train_df.ex_or_im)
co_mat_attr_ex
```

```python
# token lemma
first_word_lemma = []
for question in train_df.question:
    doc = nlp(question.lower())
    first_word_lemma.append(doc[0].lemma_)
```

```python
set(first_word_lemma)
```

```python
# question and attribute 
co_mat_qu_attr = pd.crosstab(first_word_lemma, train_df.attribute1)
co_mat_qu_attr
```

```python
# question and local_or_sum 
co_mat_qu_ls = pd.crosstab(first_word_lemma, train_df.local_or_sum)
co_mat_qu_ls
```

```python
# local_or_sum and ex_or_im
co_mat_qu_es = pd.crosstab(first_word_lemma, train_df.ex_or_im)
co_mat_qu_es
```
